```python
Price_list = []
Ratings_list=[]
Features_list=[]
Product_list=[]
Discount_list=[]
for i in range(1,10):
    url=("https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i))
    page = requests.get(url)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    list_1 = soup.find_all('div',attrs={'class':'_3pLy-c row'})
    
    for j in list_1:    
        Price=j.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
        
        if Price is None:
            Price_list.append(np.NaN)
        else:
            Price_list.append(Price.text)
            
        Features=j.find('ul',attrs={'class':'_1xgFaf'})
        if Features is None:
            Features_list.append(np.NaN)
        else:
            Features_list.append(Features.text)
        
        Ratings=j.find('div',attrs={'class':'_3LWZlK'})
        
        if Ratings is None:
            Ratings_list.append(np.NaN)
        else:
            Ratings_list.append(Ratings.text)
        
        Product=j.find('div',attrs={'class':'_4rR01T'})
        
        if Product is None:
            Product_list.append(np.NaN)
        else:
            Product_list.append(Product.text)
            
        Discount=j.find('div',attrs={'class':'_3I9_wc _27UcVY'})
        
        if Discount is None:
            Discount_list.append(np.NaN)
        else:
            Discount_list.append(Discount.text)
```

```python
df=pd.DataFrame({'Product':Product_list,'Features':Features_list,'Price':Price_list,'Ratings':Ratings_list,'Discount':Discount_list})
```
